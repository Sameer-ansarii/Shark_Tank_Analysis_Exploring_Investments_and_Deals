# Q2. Max, min , mean of asked amount,asked equity, asked valuation?

# Q13. How many times each shark invested the deal?

```python
times_each_shark_invested = df[['ashneer_deal', 'anupam_deal', 'aman_deal', 'namita_deal', 'vineeta_deal', 'peyush_deal', 'ghazal_deal']].sum()

print("Times each shark invested in the deal:\n\n", times_each_shark_invested)
```

```python
import matplotlib.pyplot as plt

shark_names = ['ashneer', 'anupam', 'aman', 'namita', 'vineeta', 'peyush', 'ghazal']
times_each_shark_invested = df[['ashneer_deal', 'anupam_deal', 'aman_deal', 'namita_deal', 'vineeta_deal', 'peyush_deal', 'ghazal_deal']].sum()

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(shark_names, times_each_shark_invested)
plt.xlabel('Shark Names')
plt.ylabel('Number of Times Invested')
plt.title('Number of Times Each Shark Invested in the Deal')

# Display the chart
plt.show()
```

# Q14. Find the equity percent that each sharks gets? 

```python
brands_with_deal = df[df['deal'] == True]['brand_name'].unique()

for brand in brands_with_deal:
    brand_data = df[df['brand_name'] == brand]
    sharks_invested = []
    equity_percentages = []
    
    for shark in ['ashneer_deal', 'anupam_deal', 'aman_deal', 'namita_deal', 'vineeta_deal', 'peyush_deal', 'ghazal_deal']:
        if brand_data[shark].any():
            sharks_invested.append(shark.split('_')[0])
            equity_percentages.append(brand_data[brand_data[shark] == True]['equity_per_shark'].values[0])
    
    print("Brand:", brand)
    print("Sharks Invested:", sharks_invested)
    print("Equity Percentages:", equity_percentages)
    print()
```

### These are the brands which gets the deal and Equity Percentage that each shark get
